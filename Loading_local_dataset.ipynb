```python
import numpy as np
import matplotlib.pyplot as plt
import os
import cv2
import random
import pickle
```

```python
DATADIR = '/home/sammie/Jupyter_notebook/observations/experiements/data'
CATEGORIES = ['with_mask', 'without_mask']

training_data = []
def creating_training_data():
    for categories in CATEGORIES:
        classification = CATEGORIES.index(categories)
        path = os.path.join(DATADIR, categories)
        for image in os.listdir(path):
            image_arr = cv2.imread(os.path.join(path, image))
            image_arr = cv2.resize(image_arr, (300, 300), cv2.COLOR_BGR2RGB)
            training_data.append([image_arr, classification])
            
creating_training_data()
```

```python
random.shuffle(training_data)
X = []
y = []
print(len(training_data))
```

```python
for imgs, labels in training_data:
    X.append(imgs)
    y.append(labels)
    
X = np.array(X).reshape(-1, 300, 300, 3)
print(X.shape)
```

```python
pickle_out = open('X.pickle', 'wb')
pickle.dump(X, pickle_out)
pickle_out.close()

pickle_out = open('y.pickle', 'wb')
pickle.dump(y, pickle_out)
pickle_out.close()
```
